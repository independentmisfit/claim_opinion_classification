```python
### YOUR CODE HERE ###
count_cols = ['video_view_count',
              'video_like_count',
              'video_share_count',
              'video_download_count',
              'video_comment_count',
              ]
for column in count_cols:
    percentile_75_view = data[column].quantile(0.75)
    percentile_25_view = data[column].quantile(0.25)
    iqr = percentile_75_view - percentile_25_view
    median = data[column].median()
    outlier_threshold = median + (1.5 * iqr)
    mask = data[column] > outlier_threshold
    num = data[column][mask].count()
    print(f'Number of outliers {column}:',num)
```
